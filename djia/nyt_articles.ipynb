```python
import pandas as pd
import glob
import os
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression, LinearRegression
```

```python
data = pd.read_csv('articles_csv/combine.csv')
data.head()
```

```python
# data['Date']
train = pd.read_csv('headlines_csv/train.csv')
test = pd.read_csv('headlines_csv/test.csv')
train.head()
```

```python
example = train.iloc[2,2]
example2 = example.lower()
```

```python
example3 = CountVectorizer().build_tokenizer()(example2)
print(example3)
```

```python
pd.DataFrame([[x,example3.count(x)] for x in set(example3)], columns = ['Word', 'Count'])
```

```python
#Get the headlines

trainheadlines = []
for row in range(0,len(train.index)):
    trainheadlines.append(' '.join(str(x) for x in train.iloc[row,2:82]))
    
trainheadlines[0]
```

```python
# Tokenizing

basicvectorizer = CountVectorizer()
basictrain = basicvectorizer.fit_transform(trainheadlines)
print(basictrain.shape)
```

```python
#Training a classifier - here Logistic

# basicmodel = LogisticRegression()
# basicmodel = basicmodel.fit(basictrain, train["Direction"])
```

```python
from sklearn.svm import LinearSVC
basicmodel = LinearSVC(fit_intercept='False')
basicmodel = basicmodel.fit(basictrain,train["Direction"])
```

```python
testheadlines = []
for row in range(0,len(test.index)):
    testheadlines.append(' '.join(str(x) for x in test.iloc[row,2:82]))
    
testheadlines[0]
```

```python
basictest = basicvectorizer.transform(testheadlines)
predictions = basicmodel.predict(basictest)
```

```python
print(predictions)
```

```python
pd.crosstab(test["Direction"], predictions, rownames=["Actual"], colnames=["Predicted"])
```
